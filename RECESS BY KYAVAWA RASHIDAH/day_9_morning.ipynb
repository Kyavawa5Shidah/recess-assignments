```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('UceResultsBySchool2011-2016.csv')
df
```

```python
#Check shape | helps to determine number of rows and columns
df.shape
```

```python
#Preparation and preprocessing or data cleaning
df.columns
```

```python
#Removing white spaces
df.columns = df.columns.to_series().apply(lambda x :x.strip())
df.columns
```

```python
#Replace column names with underscore
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
#check for information
df.info
```

```python
#Check for duplicates
for isduplicate in df.duplicated():

    if isduplicate:
        print('Duplicates identified and dropped')
        df.drop_duplicates(inplace=True)
        break
    else:
        pass
    print('No duplicates found')
    df[df.duplicated()==True]
```

```python
#Check for any null values
df.isnull().any()
```

```python
#check for the number of rows that are null
null_rows = df[df.isnull().any(axis = 1)]
before_computation = len(null_rows.axes[0])
print('Number of rows that are null values:' +str(before_computation))
```

```python
#remove rows without school names
df.dropna(thresh=None ,subset = ['SCHOOL'], inplace = True)
```
